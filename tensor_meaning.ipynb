### Tensor를 만들고 Tensor 구조와 reshape함수를 통해 변환한 Tensor의 의미를 파악해봅니다



```
%tensorflow_version 2.x
import tensorflow as tf
print(tf.version)

t = tf.zeros([5,5,5,5]) # 5*5*5*5 = 625개의 인수, arguments
print(t)

t = tf.reshape(t, [125, -1]) # -1 means automatically changed the shape according to 125
print(t)
```
